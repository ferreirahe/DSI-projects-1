```python
# Load csv file
import csv
file = []
with open("drinks.csv") as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        file.append(row)
```

```python
# Split header and data
header = file[0]
print(header)
data = file[1:]
data[15]
```

```python
# Convert data types when required
for row in data:
    for i in range(1,4):
        row[i] = int(row[i])
    row[4] = float(row[4])
```

```python
# Print sample data
data[5:8]
```

```python
# extract each column values
country = [row[0] for row in data]
beer = [row[1] for row in data]
spirit = [row[2] for row in data]
wine = [row[3] for row in data]
total = [row[4] for row in data]
```

```python
# Display statistics
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def print_stats(name,values,graph=True):
    mean = np.mean(values)
    median = np.median(values)
    min = np.min(values)
    max = np.max(values)
    std = np.std(values)
    if graph:
        plt.hist(values)
        plt.axvline(mean,color='g')
        plt.axvline(median,color="r")
        plt.show()
    print name,"consumption mean is %d" % np.floor(mean), "units, median is %d" % np.floor(median),"units"
    print "\tmin value is %d" % np.floor(min),"units and max value is %d" % np.floor(max),"units"
    print "\tstandard deviation is %f" % round(std,2)
```

```python
print_stats("beer (unit is serving)",beer)
print_stats("spirit (unit is serving)",spirit)
print_stats("wine (unit is serving)",wine)
print_stats("total (unit is liter)",total)
```

```python
def biggest_consumers(values):
    consum = [[values[i],country[i]] for i in range(len(values))]
    consum.sort(reverse=True)
    print(consum[:10])
```

```python
# What each country consumes most?
dict ={}
no_consumption =[]
beer_cons = []
spirit_cons = []
wine_cons = []
for row in data:
    max = 0
    index = 0
    for i in range(1,4):
        if row[i] > max:
            max = row[i]
            index = i
    if (index == 0):
        no_consumption.append(row[0])
    elif (index == 1):
        beer_cons.append(row[0])
    elif (index == 2):
        spirit_cons.append(row[0])
    else:
        wine_cons.append(row[0])
dict['no_cons'] = no_consumption
dict['beer'] = beer_cons
dict['spirit'] = spirit_cons
dict['wine'] = wine_cons
```

```python
# Write dicitonary to csv for later use in Tableau
with open("highest_consumption.csv",'wb') as csvfile:
    writer = csv.writer(csvfile)
    header = ["Country","Highest_consumption"]
    writer.writerow(header)
    for key,value in dict.items():
            for countr in value:
                s = [countr,key]
                writer.writerow(s)
```

```python
# Calculate amount of alcohol per unit
matrix = [[102,128,45],[193,25,221],[21,179,11]]
import numpy.linalg
result = [4.9,8.3,3.8]
factor = numpy.linalg.solve(matrix, result)
# Factor is the amount of alcohol in liter per serving
print("beer\t\tspirit\t\twine")
print(factor)
```

```python
#
beer_alc = [b*factor[0] for b in beer]
spirit_alc = [b*factor[1] for b in spirit]
wine_alc= [b*factor[2] for b in wine]
```

```python
# Calculate consumption in percentage per type
alc_in_percent = []
for i in range(len(country)):
    tot = beer_alc[i] + spirit_alc[i] + wine_alc[i]
    if tot != 0:
        percent_beer = round(100 * beer_alc[i] / tot,2)
        percent_spirit = round(100 * spirit_alc[i] / tot,2)
        percent_wine = round(100 * wine_alc[i] / tot,2)
    else:
        percent_beer=0
        percent_spirit=0
        percent_wine=0
    alc_in_percent.append([country[i],percent_beer,percent_spirit,percent_wine,])
```
